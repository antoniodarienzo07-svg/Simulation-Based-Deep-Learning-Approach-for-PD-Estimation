```python
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. GENERAZIONE DATI
np.random.seed(42)
n_customers = 2000

data = {
    'income': np.random.normal(50000, 15000, n_customers),
    'debt_ratio': np.random.uniform(0.1, 0.8, n_customers),
    'credit_score': np.random.normal(650, 50, n_customers),
    'loan_amount': np.random.uniform(5000, 40000, n_customers),
    'employment_years': np.random.poisson(5, n_customers)
}
df = pd.DataFrame(data)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
X_t = torch.tensor(X_scaled, dtype=torch.float32)

# 2. MODELLO CON LOGICA "MORBIDA" (SOFT CONSTRAINTS)
class SoftRiskNet(nn.Module):
    def __init__(self):
        super(SoftRiskNet, self).__init__()
        self.fc = nn.Linear(5, 1)
        with torch.no_grad():
            self.fc.weight.fill_(0.0)
            # --- PARAMETRO DA TOCCARE 1: SENSIBILITÀ ---
            # Abbassiamo il peso del debito da 2.5 a 1.2 per rendere il modello meno "pessimista"
            self.fc.weight[0, 1] = 1.2
            self.fc.weight[0, 2] = -0.8
            # Alziamo leggermente il bias per non avere una PD baseline troppo vicina allo zero
            self.fc.bias.fill_(-2.5)

    def forward(self, x):
        return torch.sigmoid(self.fc(x))

model = SoftRiskNet()
model.eval()

# 3. STRESS TEST CON SHOCK RIDOTTO
with torch.no_grad():
    pd_base = model(X_t).numpy().flatten()

    # --- PARAMETRO DA TOCCARE 2: INTENSITÀ DELLO SHOCK ---
    # Riduciamo lo shock da +2.0 a +0.7 deviazioni standard
    X_stress = X_t.clone()
    X_stress[:, 1] += 0.7

    pd_stress = model(X_stress).numpy().flatten()

# 4. ANALISI RISULTATI
avg_b = pd_base.mean()
avg_s = pd_stress.mean()
risk_inc = (avg_s - avg_b) / avg_b

print(f"--- MODERATED RISK REPORT ---")
print(f"Average PD (Baseline): {avg_b:.2%}")
print(f"Average PD (Stressed): {avg_s:.2%}")
print(f"Risk Increase: {risk_inc:+.2%}")

# Grafico
plt.figure(figsize=(10, 5))
plt.hist(pd_base, bins=50, alpha=0.5, label=f'Baseline (Avg: {avg_b:.2%})', color='blue')
plt.hist(pd_stress, bins=50, alpha=0.5, label=f'Stressed (Avg: {avg_s:.2%})', color='red')
plt.title('Credit Risk Migration')
plt.xlabel('Probability of Default (PD)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
